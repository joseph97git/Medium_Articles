```python
# imports for plotting
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

# imports/configurations for output
import os
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# current working directory
os.getcwd()
```

```python
# data taken from kaggle 
housing_dat = pd.read_csv('kc_house_data.csv')
```

```python
# subplot
plt.subplots(1, figsize=(12,10))

# density plot with outline  
plot_1_base = sns.kdeplot(housing_dat.sqft_living[0:100],housing_dat.price[0:100],shade=True,shade_lowest=True,cmap='plasma')
plot_1_line = sns.kdeplot(housing_dat.sqft_living[0:100],housing_dat.price[0:100],cmap='gnuplot')

# axes limits 
plot_1_base.set_xlim(0,5000)
plot_1_base.set_ylim(0,1250000)

# axes ticks
plt.xticks(list(range(0,5001,500)),fontsize=14,rotation=90.0)
plt.yticks(list(range(0,1250001, 250000)),fontsize=14)

# labels 
plt.title('Sqft_living vs Price',loc='left',size=20,va='bottom',family='monospace',color='purple',backgroundcolor='yellow')
plt.ylabel('Price',size=18,family='monospace',va='bottom',color='purple')
plt.xlabel('Sqft_living',size=18,family='monospace',va='top',color='purple')

```
